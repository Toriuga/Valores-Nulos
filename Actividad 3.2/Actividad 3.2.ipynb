```python
#importamos librerioas requeridas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Carga desde un archivo .xlsx sin indice
df= pd.read_csv('listings.csv')
df
```

```python
#hacemos una copia del archivo
data1=df.copy()

#eliminamos columnas que no deseamos 
data1= df.drop(['license','listing_url','scrape_id','source','description','neighborhood_overview','picture_url','host_url','host_location','host_about','host_thumbnail_url','host_picture_url','host_neighbourhood','host_verifications','host_has_profile_pic','neighbourhood_cleansed','neighbourhood_group_cleansed','minimum_minimum_nights','maximum_minimum_nights','minimum_maximum_nights','maximum_maximum_nights','availability_60','calendar_updated','reviews_per_month','number_of_reviews_ltm','number_of_reviews_l30d','reviews_per_month','bathrooms_text'], axis=1)
data1
```

```python
#Identificar valores nulos 
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con promedio o media
data1["bathrooms"]=data1["bathrooms"].fillna(round(data1["bathrooms"].mean(),1))
data1["bedrooms"]=data1["bedrooms"].fillna(round(data1["bedrooms"].mean(),1))
data1["beds"]=data1["beds"].fillna(round(data1["beds"].mean(),1))

#Sustituir valores nulos por un string en concreto
data1["host_response_time"] =data1["host_response_time"].fillna("within an hour")
data1["host_response_rate"] =data1["host_response_rate"].fillna("50%")
data1["host_acceptance_rate"] =data1["host_acceptance_rate"].fillna("70%")
data1["host_is_superhost"] =data1["host_is_superhost"].fillna("f")
data1["has_availability"] =data1["has_availability"].fillna("t")
data1["first_review"] =data1["first_review"].fillna("2030-06-19")
data1["last_review"] =data1["last_review"].fillna("2030-06-19")

#Sustituir valores nulos por valores no nulos hacia adelante "forward fill"
data1["neighbourhood"] =data1["neighbourhood"].fillna(method="ffill")
#Sustituir valores nulos por valores no nulos hacia atrás backward fill"("bfill")
data1["neighbourhood"] =data1["neighbourhood"].fillna(method="bfill")

#sustituir valores nulos con mediana
data1["review_scores_rating"]=data1["review_scores_rating"].fillna(round(data1["review_scores_rating"].median(),1))
data1["review_scores_accuracy"]=data1["review_scores_accuracy"].fillna(round(data1["review_scores_accuracy"].median(),1))
data1["review_scores_cleanliness"]=data1["review_scores_cleanliness"].fillna(round(data1["review_scores_cleanliness"].median(),1))
data1["review_scores_checkin"]=data1["review_scores_checkin"].fillna(round(data1["review_scores_checkin"].median(),1))
data1["review_scores_communication"]=data1["review_scores_communication"].fillna(round(data1["review_scores_communication"].median(),1))
data1["review_scores_location"]=data1["review_scores_location"].fillna(round(data1["review_scores_location"].median(),1))
data1["review_scores_value"]=data1["review_scores_value"].fillna(round(data1["review_scores_value"].median(),1))

# 1. Eliminar el símbolo de dólar y convertir a tipo numérico
data1["price"] = data1["price"].replace('[\$,]', '', regex=True).astype(float)
# 2. Calcular la mediana de la columna
mediana = round(data1["price"].median(), 1)
# 3. Sustituir los valores nulos con la mediana
data1["price"] = data1["price"].fillna(mediana)

#Identificar valores nulos 
valores_nulos=data1.isnull().sum()
valores_nulos


#Convertir archivo filtrado a CSV
data1.to_csv("Actividad 3.2.csv")

```
